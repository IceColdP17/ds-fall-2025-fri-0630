## Dataset Setup

First, let's generate the synthetic dataset you'll be analyzing:

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Set random seed for reproducible results
np.random.seed(42)

# Generate 500 orders from 2023
n_orders = 500

# Create order IDs and customer data
order_id = np.arange(1, n_orders + 1)
customer_id = np.random.randint(1, 101, size=n_orders)
customer_segment = np.random.choice(["Student", "Professional", "Senior"], size=n_orders, p=[0.4, 0.4, 0.2])
region = np.random.choice(["North", "South", "East", "West"], size=n_orders)
order_date = pd.date_range(start="2023-01-01", periods=n_orders, freq="D")
product_category = np.random.choice(["Electronics", "Clothing", "Books", "Home", "Sports"], size=n_orders)
quantity = np.random.randint(1, 5, size=n_orders)
unit_price = np.round(np.random.uniform(10, 200, size=n_orders), 2)
revenue = quantity * unit_price
discount_applied = np.random.choice([0, 5, 10, 15, 20, 25, 30], size=n_orders, p=[0.5,0.1,0.1,0.1,0.05,0.05,0.1])
payment_method = np.random.choice(["Credit Card", "PayPal", "Gift Card", "Debit Card"], size=n_orders)

# Build dataframe
df = pd.DataFrame({
    "order_id": order_id,
    "customer_id": customer_id,
    "customer_segment": customer_segment,
    "region": region,
    "order_date": order_date,
    "product_category": product_category,
    "quantity": quantity,
    "unit_price": unit_price,
    "revenue": revenue,
    "discount_applied": discount_applied,
    "payment_method": payment_method
})

# Save dataset for reference
df.to_csv("ecommerce_data.csv", index=False)
print("✅ Dataset created successfully!")

# Display first few rows
df.head()
```

## Data Exploration

Let's start by understanding our dataset:

```python
# Load the data
data = pd.read_csv("ecommerce_data.csv")

# Basic information about the dataset
print("Dataset Info:")
print(data.info())
print("\nDataset Shape:", data.shape)
print("\nFirst few rows:")
data.head()
```

```python
# Convert order_date to datetime and add helpful columns
data["order_date"] = pd.to_datetime(data["order_date"])
data["month"] = data["order_date"].dt.to_period("M")
data["day_of_week"] = data["order_date"].dt.day_name()

# Example: Revenue by month
monthly_revenue = data.groupby("month")["revenue"].sum()
print("Monthly Revenue:")
print(monthly_revenue)

# Plot monthly revenue
monthly_revenue.plot(kind="bar", figsize=(10,6), title="Monthly Revenue")
plt.ylabel("Revenue ($)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 1. Revenue Overview
What was the company's total revenue last quarter? Break it down by month.

```python
final_quarter = [10, 11, 12]
month_dict = { 1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'April', 5: 'May', 6: 'Jun', 7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct',11: 'Nov', 12: 'Dec'}
final_quarter_df = data[data['month'].dt.month.isin(final_quarter)].copy()
final_quarter_earnings = final_quarter_df['revenue'].sum()
print(f'|{"-"*42}|\n| Total Income for Last Quarter: ${final_quarter_earnings} |\n|{"-"*42}|')

final_quarter_monthly_earnings = final_quarter_df.groupby('month')['revenue'].sum().reset_index()
print(f'| Monthly Earnings:{" "*24}|\n|{" "*42}|')
for index, row in final_quarter_monthly_earnings.iterrows():
    print(f'|{" "*6}{"="*11}{" "*25}|\n|{" "*5}{month_dict[row["month"].month]}: ${row["revenue"]:.2f}{" "*24}|')
print(f'|{"-"*42}|')
```

### 2. Top Customers
Who are the top 5 customers by total revenue contribution?

```python
# Your code here
top_5_customers = data.groupby('customer_id')['revenue'].sum().reset_index()
top_5_customers = top_5_customers.sort_values('revenue', ascending=False).head(5).reset_index(drop=True)
top_5_customers
```

### 3. Segment Analysis
Which customer segment (Student, Professional, Senior) spends the most on average per order?

```python
# Your code here
top_segment = data.groupby('customer_segment')['revenue'].mean().head(1)
top_segment
```

### 4. Regional Trends
Which region generated the highest revenue? Which region had the highest average order size?

```python
# Your code here
highest_regional_rev = data.groupby('region')["revenue"].sum().sort_values(ascending=False).head(1).reset_index()
region_name = highest_regional_rev.loc[0, "region"]
region_revenue = highest_regional_rev.loc[0, "revenue"]
print(f"Highest Grossing Region by Revenue: {region_name} (${region_revenue:,.2f})")

highest_regional_average = data.groupby('region')["revenue"].mean().sort_values(ascending=False).head(1).reset_index()
avg_region_name = highest_regional_average.loc[0, "region"]
avg_order_size = highest_regional_average.loc[0, "revenue"]
print(f"Region with Highest Average Order Size: {avg_region_name} (${avg_order_size:,.2f} per order)")
```

### 5. Product Category Performance
Which product category is most popular by quantity vs. by revenue?

```python
# Your code here
most_popular_prod_by_quantity = data.groupby('product_category')['quantity'].sum().sort_values(ascending=False).head(1).reset_index()
most_popular_category = most_popular_prod_by_quantity.loc[0, "product_category"]
most_popular_category_quantity = most_popular_prod_by_quantity.loc[0, "quantity"]
print(f'Most Popular Product Category by Quantity: {most_popular_category} - {most_popular_category_quantity} sold')

most_popular_prod_by_revenue = data.groupby('product_category')['revenue'].sum().sort_values(ascending=False).head(1).reset_index()
most_popular_category_by_rev = most_popular_prod_by_revenue.loc[0, "product_category"]
most_popular_category_rev = most_popular_prod_by_revenue.loc[0, "revenue"]
print(f'Most Popular Product Category by Revenue: {most_popular_category_by_rev} - ${most_popular_category_rev:.2f} generated')
```

### 6. Discount Effectiveness
Do discounted orders generate higher or lower revenue per order compared to non-discounted orders?


```python
# Your code here
discounted_sales_df = data[data['discount_applied'] > 0]
regular_sales_df = data[data['discount_applied'] == 0]

mean_discounted_sales = discounted_sales_df['revenue'].mean()
mean_regular_sales = regular_sales_df['revenue'].mean()

print(f'Discounted Sales generate more revenue at ${mean_discounted_sales:.2f} per order' \
      if mean_discounted_sales > mean_regular_sales else \
        f'Regular Sales generate more revenue at ${mean_regular_sales:.2f}')
```

### 7. Payment Method Usage
What percentage of orders use each payment method? Does any payment method correlate with higher spending?

```python
# Your code here
payment_percentages_per_method = (data['payment_method'].value_counts(normalize=True) * 100).reset_index()
payment_percentages_per_method.columns = ['payment_method', 'percentage']
for index, row in payment_percentages_per_method.iterrows():
    print(f'{row["payment_method"]}: {row["percentage"]:.2f}% of orders')

dummy_payments = pd.get_dummies(data['payment_method'])
payment_methods_vs_revenue_corr = pd.concat([data['revenue'], dummy_payments], axis=1).corr()
plt.figure(figsize=(8, 5))
sns.heatmap(payment_methods_vs_revenue_corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap: Revenue vs Payment Methods')
plt.show()
```

The heatmap shows that Debit Cards is the payment method most correlated with higher spending

### 8. Seasonality Check
Plot revenue by day of week — are weekends busier than weekdays?

```python
# Your code here
weekly_spending = data.groupby('day_of_week')['revenue'].sum().reset_index()
days_ordered = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
weekly_spending['day_of_week'] = pd.Categorical(weekly_spending['day_of_week'], categories=days_ordered, ordered=True)
weekly_spending = weekly_spending.sort_values('day_of_week')

plt.figure(figsize=(8,5))
sns.barplot(weekly_spending, x='day_of_week', y='revenue')
```

Using the graph as evidence, it seems that weekends are not busier than weekdays, in fact the two busiest days are on Tuesdays and Wednesdays, althougth they are not busier than the other days by a significant margin

### 9. Profitability (Stretch)
Assume profit margin = 20% of revenue. Estimate total profit and identify the most profitable region.

```python
# Your code here
data['profit'] = data['revenue'] * 0.20
total_profit = data['profit'].sum()

regional_profit = data.groupby('region')['profit'].sum().reset_index()
most_profitable_region = regional_profit.loc[regional_profit['profit'].idxmax(), 'region']
most_profitable_profit = regional_profit['profit'].max()
print(f"Total Profit: ${total_profit:,.2f}")
print(f"Most Profitable Region: {most_profitable_region}ern region - ${most_profitable_profit:,.2f}")
regional_profit
```
