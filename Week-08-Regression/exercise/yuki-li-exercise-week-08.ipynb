````python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, f1_score

df.head(5)
````

````python
dependent_var = 'PTS'
independent_var = ['2PA', '3PA', 'FTA', 'ORB', 'AST']

X = df[independent_var]
y = df[dependent_var]

print(f"Target variable: {dependent_var}")
print(f"Features: {independent_var}")
print(f"Sample size: {len(X)} observations")
print("\nFirst 5 rows of features:")
X.head()
````

````python
X_with_constant = sm.add_constant(X)

model_v1 = sm.OLS(y, X_with_constant).fit()

y_pred = model_v1.predict(X_with_constant)

print("Initial model built with all available features")
print(f"Model R-squared: {model_v1.rsquared:.4f}")
print(f"Number of features: {len(independent_var)}") 
````

````python
# Display the comprehensive model summary
print("="*80)
print("REGRESSION MODEL SUMMARY")
print("="*80)
results = model_v1.summary2()
print(results)

# Calculate and display key performance metrics
rmse_score = rmse(y, y_pred)
print("\n" + "="*80)
print("MODEL PERFORMANCE METRICS")
print("="*80)
print(f"R-squared: {model_v1.rsquared:.4f}")
print(f"Adjusted R-squared: {model_v1.rsquared_adj:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse_score:.4f}")
print(f"Mean of dependent variable ({dependent_var}): {y.mean():.2f}")
print(f"RMSE as % of mean: {(rmse_score / y.mean()) * 100:.2f}%")
````

````python
# =============================================================================
# EXPLORATORY DATA ANALYSIS: RELATIONSHIPS WITH RUNS SCORED
# =============================================================================

# Create comprehensive scatter plots to visualize relationships
dependent_var = 'PTS'
independent_var = ['2PA', '3PA', 'FTA', 'ORB', 'AST']

# Set up the the number of grids needed
n_vars = len(independent_var)
cols = 2
rows = (n_vars + cols - 1) // cols

# Create figure with subplots
fig, axes = plt.subplots(rows, cols, figsize=(6 * cols, 5 * rows))
axes = axes.flatten()  # Flatten for easy iteration

print("Creating scatter plots to examine relationships with Runs Scored...")

# Generate scatter plots for each variable
for i, var in enumerate(independent_var):
    # Create scatter plot with regression line
    sns.scatterplot(x=df[var], y=df[dependent_var], 
                   ax=axes[i], alpha=0.6, s=50)
    
    # Add regression line to show trend
    sns.regplot(x=df[var], y=df[dependent_var], 
               ax=axes[i], scatter=False, color='red', line_kws={'linewidth': 2})
    
    # Customize plot appearance
    axes[i].set_title(f'{var} vs {dependent_var}', fontsize=12, fontweight='bold')
    axes[i].set_xlabel(var, fontsize=10)
    axes[i].set_ylabel(dependent_var, fontsize=10)
    axes[i].grid(True, alpha=0.3)

# Hide any empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.suptitle('Exploratory Analysis: Variables vs Points', 
             fontsize=16, fontweight='bold', y=1.02)
plt.show()

print("Analysis: Look for strong linear relationships and potential outliers")
````

Evaluate if there are any multicollinearity

````python
# =============================================================================
# MULTICOLLINEARITY ANALYSIS: CORRELATION MATRIX
# =============================================================================

print("="*80)
print("CORRELATION ANALYSIS")
print("="*80)

# Calculate correlation matrix for our offensive variables
correlation_matrix = df[independent_var].corr().round(4)

print("Correlation Matrix:")
print(correlation_matrix)

# Create a heatmap for visual analysis
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
            square=True, fmt='.3f', cbar_kws={'shrink': 0.8})
plt.title('Correlation Matrix: Offensive Statistics', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
````
